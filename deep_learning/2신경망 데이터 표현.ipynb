# 신경망 데이터 표현

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['seaborn-whitegrid'])
```

## 텐서(Tensor)
* 일반적으로 텐서는 3차원 이상을 다룰 때 표현하는 방식이지만, 여기서는 어떠한 데이터를 표현할 때,그 값 모두를 텐서라고 부르기로 함
* 
* a(1차원),b(2차원),c(1차원),d(0차원-스칼라) 모두 텐서라고 지칭할 수 있음
* 랭크(rank) : 텐서의 축을 나타내고, 넘파이의 ndim으로 구현할 수 있음
* 
    * 위의 예시에서 알 수 있듯이 대괄호([])의 개수가 곧 랭크의 값
* 크기 (shape) : 텐서의 각 축을 따라 얼마나 많은 차원이 있는지를 나타내며, 파이썬의 튜플 형태

## 그림으로 이해하기


## 스칼라 - 0차원 텐서
* 하나의 숫자를 담고 있는 텐서
* 형상은 없음

```python
x = np.array(3)
print(x)
print(x.shape) 
print(np.ndim(x))
```

## 벡터 - 1차원 텐서
* 숫자의 배열을 나타내는 텐서

```python
x = np.array([1,2,3,4])
print(x)
print(x.shape) # vector 1차원 텐서
print(np.ndim(x))
```

## 벡터의 합
* 같은 형상(shape)일때, 각 원소별로 계산

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])

c = a+b
print(c) # 같은 자리의 원소끼리 + 연산
print(c.shape) # shape은 그대로 1차원 텐서
print(np.ndim(c))
```

## 벡터의 곱
* 
* 원소곱
    * 같은 shape일때, 각 원소별로 계산
    
* 벡터곱 (product,dot)
    * 두 1차원 벡터가 있을 때, 각각의 성분끼리의 곱을 모두 더하는 계산 -> 스칼라값
    

```python
a = np.array([1,2,3,4])
b = np.array([5,6,7,8])

c = a*b
print(c)
print(c.shape)
print(np.ndim(c))
print('--------------')
x = np.array([1,2,0])
y = np.array([0,2,1])
z = np.dot(x,y)
print(z)
print(z.shape)
print(np.ndim(z))
```

## 스칼라와 벡터의 곱

```python
a = np.array(10)
b = np.array([1,2,3])
print(a*b)
```

## 2차원 텐서 (행렬)
* 2차원 텐서는 행렬로 생각할 수 있음
    * (m,n)형상의 배열
    

```python
matrix = np.array([[1,2,3]
                  ,[4,5,6]])

print(matrix)
print(matrix.shape)
print(np.ndim(matrix))
print('------------------')
matrix2 = np.array([[1,2,3,4]])
print(matrix2)
print(matrix2.shape)
print(np.ndim(matrix2))
```

## 행렬 원소곱
* 같은 shape일 때 덧셈, 곱셈과 같은 연산은 "원소별"로 진행

```python
A = np.array([[1,2],[3,4]])
B = np.array([[10,10],[10,10]])

print('행렬 A\n',A)
print('행렬 B\n',B)
print('A*B\n',A*B)
```

## 행렬 점곱 (product)
* 1차원 벡터와 마찬가지로 앞 행렬의 열과 뒤행렬의 행의 수가 같아야함

```python
M = np.array([[1,2],[3,4]])
N = np.array([[2,3,4],[2,3,4]])

print('행렬 M\n',M)
print('행렬 N\n',N)
L = np.dot(M,N)
print('행렬 L\n',L)
print(L.shape)
print(np.ndim(L))
```

```python
m = np.array([[1],[2],[3]]) # 3x1
n = np.array([[1],[2],[3]]) # 3x1

l = np.dot(m,n)

print(l)
print(l.shape)
print(np.ndim(l))
# error가 발생한다.
```

## 역행렬
* 어떤 행렬 A가 있을 때, 곱해서 단위행렬을 만드는 행렬 B가 존재한다면, 행렬 B는 A의 역행렬

```python
A = np.array([[1,2],[3,4]])
print(A)
B = np.linalg.inv(A)
print(B)
print(A*B)
print(np.dot(A,B)) #단위행렬이 나온다.
```

```python
B = np.array([3,3],[2,2])
print(np.linalg.inv(B)) #역행렬이 존재하지 않는 행렬 ad-bc=0
```

## 전치행렬
* 행과 열을 바꾼 형태
* 

```python
A = np.array([[1,2,3],[4,5,6]])
print('A\n',A)
print('A.shape\n',A.shape)
print('--------------------')
A_ = A.T
print('A의 전치행렬\n',A_)
print('A.T.shape\n',A_.shape)
```

## 3차원 텐서
* 보통 이미지를 나타낼 때 사용되는 텐서
    * (width,height,channels) : channels -> RGB값
    * 일반적으로 Numpy array로 표현
    * 
* 시계열데이터 또는 시퀀스 데이터를 표현할 때도 사용

## 브로드캐스팅(broadcasting)
* 넘파이에서 다른 shape끼리 계산이 가능하도록 
* 1차원 텐서 -> 각 원소에 더해준다.
* 
* 2차원 텐서 -> 부족한부분을 채워서
* 
* (참고) 아래의 경우도 가능
* 

```python
a = np.array(10)
b = np.array([10,20,30])
print(np.dot(a,b))
print(a*b)
```

```python
A = np.array([[1,2],[3,4]])
B = np.array([10,20])

print('행렬 A\n',A)
print('행렬 B\n',B)
print('A * B\n',A*B)
```

```python
A = np.array([[[1,1,1],
              [2,2,2]],
              [[3,3,3],
             [4,4,4]]])
B = np.array([[10,10,10]])

print('행렬 A\n',A)
print('A.shape',A.shape)
print('행렬 B\n',B)
print('B.shape',B.shape)
print('A * B\n',A*B)
```

```python
A = np.array([[1,2,3],[4,5,6]])
B = np.array([10,10])

print(A*B) #브로드캐스팅 조건이 맞지 않는다.
```

## 텐서 크기 변환
* reshape로 텐서의 크기 변환 가능
* 변환전의 원소의 개수와 변환 이후의 텐서의 개수가 같아야함

```python
A = np.array([[1,2,3],[4,5,6]])
print('행렬 A\n',A)
print('A.shape:',A.shape)
print('-----------------')
A = A.reshape(6)
print('행렬 A\n',A)
print('A.shape:',A.shape)
```

```python
B = np.array([[[2,3,4],[2,3,4]],
             [[1,1,1],[1,1,1]]])
print('행렬 B\n',B)
print('B.shape:',B.shape)
print('-------------')
B = B.reshape(3,4)
print('행렬 B\n',B)
print('B.shape:',B.shape)
```

* -1을 통해 자동으로 형상을 지정가능 <br>
원소의 개수에 맞게 넘파이가 자동으로 형상을 지정

```python
(2,2,3) -> (3,-1)
        -> (2,1,6)
        -> (2,-1,-1) x
        -> (2,5,-1)  x
```

```python
B = np.array([[[2,3,4],[2,3,4]],
             [[1,1,1],[1,1,1]]])
print('행렬 B\n',B)
print('B.shape:',B.shape)
print('-------------')
B = B.reshape(4,-1)
print('행렬 B\n',B)
print('B.shape:',B.shape)
```
