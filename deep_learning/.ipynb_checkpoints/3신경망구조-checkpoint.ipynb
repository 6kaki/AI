## AND 
*    1  1  
    * 
* 
    * 
* AND    

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def AND(a,b):
    input = np.array([a,b])
    weights  = np.array([0.4,0.4])
    bias = -0.6
    value = np.sum(input * weights) + bias
    
    if value <= 0:
        return 0
    else:
        return 1
```

```python
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

```python
x1 = np.arange(-2,2,0.01)
x2 = np.arange(-2,2,0.01)
bias = -0.6

y = (-0.4 * x1 - bias) / 0.4

plt.axvline(x=0)
plt.axhline(y=0)
plt.plot(x1,y,'r--')
# plt.plot(x2,y,'k--')   .
plt.scatter(0,0,color='orange',marker='o',s=150)
plt.scatter(1,0,color='orange',marker='o',s=150)
plt.scatter(0,1,color='orange',marker='o',s=150)
plt.scatter(1,1,color='black',marker='^',s=150)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
plt.show()
```

## OR
*     1 1  
    * 
* 
    * 
* OR    

```python
def OR(a,b):
    input = np.array([a,b])
    weights = np.array([0.4,0.5])
    bias = -0.3
    value = np.sum(input * weights) + bias
    
    if value <= 0:
        return 0
    else:
        return 1
```

```python
print(OR(0,0))
print(OR(1,0))
print(OR(0,1))
print(OR(1,1))
```

```python
x1 = np.arange(-2,2,0.01)
x2 = np.arange(-2,2,0.01)
bias = -0.3

y = (-0.4 * x1 - bias) / 0.5

plt.axvline(x=0)
plt.axhline(y=0)
plt.plot(x1,y,'r--')
# plt.plot(x2,y,'k--')   .
plt.scatter(0,0,color='black',marker='^',s=150)
plt.scatter(1,0,color='orange',marker='o',s=150)
plt.scatter(0,1,color='orange',marker='o',s=150)
plt.scatter(1,1,color='orange',marker='o',s=150)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
plt.show()
```

## NAND 
*    1  0  
    * 
* 
    * 

```python
def NAND(a,b):
    inputs = np.array([a,b])
    weights = np.array([-0.6,-0.5])
    bias = 0.7
    value = np.sum(inputs * weights) + bias
    
    if value <= 0:
        return 0
    else:
        return 1
```

```python
print(NAND(0,0))
print(NAND(1,0))
print(NAND(0,1))
print(NAND(1,1))
```

```python
x1 = np.arange(-2,2,0.01)
x2 = np.arange(-2,2,0.01)
bias = 0.7

y = (0.6 * x1 - bias) / -0.5

plt.axvline(x=0)
plt.axhline(y=0)
plt.plot(x1,y,'r--')
# plt.plot(x2,y,'k--')   .
plt.scatter(0,0,color='orange',marker='o',s=150)
plt.scatter(1,0,color='orange',marker='o',s=150)
plt.scatter(0,1,color='orange',marker='o',s=150)
plt.scatter(1,1,color='black',marker='^',s=150)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
plt.show()
```

*  (hidden layer)
    * 1

*  (output layer)
<br>

## XOR
*       1 
* 
    * 

```python
def XOR(x1,x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)   # OR NAND Layer    -> hidden layer
    y = AND(s1,s2)
    return y
```

```python
print(XOR(0,0))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(1,1))
```

## Step Function( )

```python
def step_function(x):
    if x>=0:
        return 1
    else:
        return 0
```

```python
def step_function_for_numpy(x):
    y = x>0 # y  int 
    return y.astype(np.int64)
```

## Hyperbolic tangent function) tanh
* 
* 
* sigmoid  0~1
* tanh -1~1

```python
def tanh(x):
    return ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
```

```python
print(tanh(3))
print(tanh(-3))
```

## Softmax function
*     (Multi-class Classification)
*    
    *    
*   
    *   1
* 
    * 
* 

```python
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
a = np.array([0.3,0.2,0.4,-1.2])
print(softmax(a)) # .
print(np.sum(softmax(a))) # 1
```

## softmax  
*  (overflow)
*  (exponential function)       
*  
    * 

```python
A = np.array([1000,900,1050,500])
print(softmax(A)) #  
```

```python
def softmax_(a):
    C = np.max(a) #scaling
    return (np.exp(a-C) / np.sum(np.exp(a-C)))
```

```python
A = np.array([1000,900,1050,500])
print(softmax_(A)) #  
```

* Leaky ReLU
* 
* 
*  0    

```python
def Leaky_ReLU(x):
    a = 0.01
    return np.maximum(a*x,x)
```

```python
x = np.array([0.5,-1.4,3.0,0,5])
print(Leaky_ReLU(x))
print(x)
print(0.01*x)
```

* ELU (exponential linear units)
* 
*   

```python
def elu(x):
    alpha = 1.0
    return(x>=0)*x + (x<0) * alpha * (np.exp(x)-1)
```

```python
print(elu(4))
print(elu(-0.5))

x = np.array([-2,0.1,4])
print(elu(x))
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

###  

```python
X = np.array([1.0,0.5,0.4])
W1 = np.array([[0.1,0.3,0.5],[0.2,0.4,0.6],[0.3,0.5,0.7]])
B1 = np.array([1,1,1])

print(W1.shape)
print(X.shape)
print(B1.shape)

A1 = np.dot(X,W1) + B1
Z1 = sigmoid(A1)
print(A1)
print(Z1)
```

```python
W2 = np.array([[0.2,0.4,0.6],[0.1,0.3,0.5],[0.4,0.6,0.8]])
B2 = np.array([1,1,1])

print(W2.shape)
print(B2.shape)

A2 = np.dot(Z1,W2) + B2
Z2 = sigmoid(A2)

print(A2)
print(Z2)
```

```python
W3 = np.array([[0.1,0.3],[-0.1,-0.5],[0.3,0.5]])
B3 = np.array([1,1])

print(W3.shape)
print(B3.shape)

A3 = np.dot(Z2,W3) + B3
Z3 = sigmoid(A3)

print(A3)
print(Z3)
```

```python
W4 = np.array([[0.1,0.2],[0.3,0.5]])
B4 = np.array([1,1])

print(W4.shape)
print(B4.shape)

A4 = np.dot(Z3,W4) + B4
Y = sigmoid(A4)

print(A4)
print(Y)
```

```python
def network():
    network = {}
    # 
    network['W1'] = np.array([[0.1,0.3,0.5],[0.2,0.4,0.6],[0.3,0.5,0.7]])
    network['B1'] = np.array([1,1,1])
    
    # 
    network['W2'] = np.array([[0.2,0.4,0.6],[0.1,0.3,0.5],[0.4,0.6,0.8]])
    network['B2'] = np.array([1,1,1])
    
    # 
    network['W3'] = np.array([[0.1,0.3],[-0.1,-0.5],[0.3,0.5]])
    network['B3'] = np.array([1,1])
    
    # 
    network['W4'] = np.array([[0.1,0.2],[0.3,0.5]])
    network['B4'] = np.array([1,1])
    
    return network
```

```python
def forward(network,x):
    W1,W2,W3,W4 = network['W1'],network['W2'],network['W3'],network['W4']
    B1,B2,B3,B4 = network['B1'],network['B2'],network['B3'],network['B4']
    
    A1 = np.dot(x,W1) + B1
    Z1 = sigmoid(A1)  
    
    A2 = np.dot(Z1,W2) + B2
    Z2 = sigmoid(A2)
    
    A3 = np.dot(Z2,W3) + B3
    Z3 = sigmoid(A3)
        
    A4 = np.dot(Z3,W4) + B4
    Y = sigmoid(A4)
    
    return Y
```

##   

```python
net = network()
x = np.array([0.3,1.3,-2.2])
y = forward(net,x)

print(y)
```
