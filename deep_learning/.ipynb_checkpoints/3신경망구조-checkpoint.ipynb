## AND 게이트
* 두 입력이 모두 1일 때 1을 출력하는 논리회로
    * 
* 진리표
    * 
* AND게이트를 만족시키는 가중치와 편향 구하기

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def AND(a,b):
    input = np.array([a,b])
    weights  = np.array([0.4,0.4])
    bias = -0.6
    value = np.sum(input * weights) + bias
    
    if value <= 0:
        return 0
    else:
        return 1
```

```python
print(AND(0,0))
print(AND(1,0))
print(AND(0,1))
print(AND(1,1))
```

```python
x1 = np.arange(-2,2,0.01)
x2 = np.arange(-2,2,0.01)
bias = -0.6

y = (-0.4 * x1 - bias) / 0.4

plt.axvline(x=0)
plt.axhline(y=0)
plt.plot(x1,y,'r--')
# plt.plot(x2,y,'k--') 위와 똑같이 그려진다.
plt.scatter(0,0,color='orange',marker='o',s=150)
plt.scatter(1,0,color='orange',marker='o',s=150)
plt.scatter(0,1,color='orange',marker='o',s=150)
plt.scatter(1,1,color='black',marker='^',s=150)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
plt.show()
```

## OR게이트
* 두 입력 중 하나라도 1이면 1을 출력하는 논리회로
    * 
* 진리표
    * 
* OR게이트를 만족시키는 가중치와 편향 구하기

```python
def OR(a,b):
    input = np.array([a,b])
    weights = np.array([0.4,0.5])
    bias = -0.3
    value = np.sum(input * weights) + bias
    
    if value <= 0:
        return 0
    else:
        return 1
```

```python
print(OR(0,0))
print(OR(1,0))
print(OR(0,1))
print(OR(1,1))
```

```python
x1 = np.arange(-2,2,0.01)
x2 = np.arange(-2,2,0.01)
bias = -0.3

y = (-0.4 * x1 - bias) / 0.5

plt.axvline(x=0)
plt.axhline(y=0)
plt.plot(x1,y,'r--')
# plt.plot(x2,y,'k--') 위와 똑같이 그려진다.
plt.scatter(0,0,color='black',marker='^',s=150)
plt.scatter(1,0,color='orange',marker='o',s=150)
plt.scatter(0,1,color='orange',marker='o',s=150)
plt.scatter(1,1,color='orange',marker='o',s=150)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
plt.show()
```

## NAND 게이트
* 두 입력이 모두 1일 때 0을 출력하는 논리회로
    * 
* 진리표
    * 

```python
def NAND(a,b):
    inputs = np.array([a,b])
    weights = np.array([-0.6,-0.5])
    bias = 0.7
    value = np.sum(inputs * weights) + bias
    
    if value <= 0:
        return 0
    else:
        return 1
```

```python
print(NAND(0,0))
print(NAND(1,0))
print(NAND(0,1))
print(NAND(1,1))
```

```python
x1 = np.arange(-2,2,0.01)
x2 = np.arange(-2,2,0.01)
bias = 0.7

y = (0.6 * x1 - bias) / -0.5

plt.axvline(x=0)
plt.axhline(y=0)
plt.plot(x1,y,'r--')
# plt.plot(x2,y,'k--') 위와 똑같이 그려진다.
plt.scatter(0,0,color='orange',marker='o',s=150)
plt.scatter(1,0,color='orange',marker='o',s=150)
plt.scatter(0,1,color='orange',marker='o',s=150)
plt.scatter(1,1,color='black',marker='^',s=150)
plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.grid()
plt.show()
```

## XOR게이트
* 서로 다른 두 값이 입력으로 들어가면 1을 반환
* 진리표
    * 

```python
def XOR(x1,x2):
    s1 = NAND(x1,x2)
    s2 = OR(x1,x2)   # OR과 NAND라는 Layer를 두고 출력한 것 -> hidden layer
    y = AND(s1,s2)
    return y
```

```python
print(XOR(0,0))
print(XOR(1,0))
print(XOR(0,1))
print(XOR(1,1))
```

## Step Function(계단 함수)
* 

```python
def step_function(x):
    if x>=0:
        return 1
    else:
        return 0
```

```python
def step_function_for_numpy(x):
    y = x>0 #참거짓을 y에 저장하고 int값으로 반환
    return y.astype(np.int64)
```

```python
print(step_function(-3))
print(step_function(5))
print()
a = np.array([5,3,-4,2.0])
print(step_function_for_numpy(a))
```

## Sigmoid 함수
* 이진분류(binary  classification)에 주로 사용
    * 마지막 출력층의 활성화 함수로 사용
* 출력값이 0~1의 값이며, 이는 확률로 표현가능
* 
* 
* 0.5를 기준으로 

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

```python
print(sigmoid(3))
print(sigmoid(-3))
```

## sigmoid함수와 step함수 비교
* 공통점
    * 출력값이 0~1내의 범위
    * 입력값의 정도에 따라 출력값의 정도가 달라짐 즉, 입력이 중요하면(입력값이 크면) 큰값을 출력
* 차이점
    * 계단함수에 비해 시그모이드함수는
        * 입력에 따라 출력이 연속적으로 변화
        * 출력이 '매끄러움' -> 이는 모든 점에서 미분가능함을 의미

```python
plt.grid()
x = np.arange(-5.0,5.0,0.01)
y1 = sigmoid(x)
y2 = step_function_for_numpy(x)
plt.plot(x,y1,'r-',x,y2,'b--')
plt.show()
```

## Hyperbolic tangent function) tanh
* 
* 
* sigmoid와 비슷하지만 0~1
* tanh는 -1~1까지

```python
def tanh(x):
    return ((np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)))
```

```python
print(tanh(3))
print(tanh(-3))
```

## Softmax function
* 다중 클래스 분류에서 사용 (Multi-class Classification)
* 입력값의 영향을 크게 받음
    * 입력값이 크면 출력값도 큼
* 출력값을 확률에 대응가능
    * 출력값의 총합은 1
* 수식
    * 
* 

```python
def softmax(a):
    exp_a = np.exp(a)
    sum_exp_a = np.sum(exp_a)
    y = exp_a / sum_exp_a
    return y
```

```python
a = np.array([0.3,0.2,0.4,-1.2])
print(softmax(a)) #확률값으로 나온다.
print(np.sum(softmax(a))) #총합은 1
```

## softmax 함수 주의점
* 오버플로우 (overflow)문제
* 지수함수 (exponential function)을 사용하기 때문에 입력값이 너무 크면 무한대가 반환됨
* 개선한 수식
    * 

```python
A = np.array([1000,900,1050,500])
print(softmax(A)) # 오버플로우 발생
```

```python
def softmax_(a):
    C = np.max(a) #scaling
    return (np.exp(a-C) / np.sum(np.exp(a-C)))
```

```python
A = np.array([1000,900,1050,500])
print(softmax_(A)) # 오버플로우 발생안함
```

    * 
    * 결국 선형함수이다.

## 그 외의 활성화 함수
* Leaky ReLU
* 
* 
* 그냥 0으로 하지 않고 약간 줄어들도록

```python
def Leaky_ReLU(x):
    a = 0.01
    return np.maximum(a*x,x)
```

```python
x = np.array([0.5,-1.4,3.0,0,5])
print(Leaky_ReLU(x))
print(x)
print(0.01*x)
```

* ELU (exponential linear units)
* 
* 매끄럽게 굴곡을 줘서

```python
def elu(x):
    alpha = 1.0
    return(x>=0)*x + (x<0) * alpha * (np.exp(x)-1)
```

```python
print(elu(4))
print(elu(-0.5))

x = np.array([-2,0.1,4])
print(elu(x))
```

## 3층 신경망 구현하기
* 
* 2클래스 분류
* 입력층 (input layer)
    * 뉴런수 : 3
* 은닉층  (hidden layer)
    * 첫번째 은닉층
        * 뉴런수 : 3
    * 두번째 은닉층
        * 뉴런수 : 2
* 출력층 (output layer)
    * 뉴런수 : 2

### 활성화 함수 정의

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

### 레이어 정의

```python
X = np.array([1.0,0.5,0.4])
W1 = np.array([[0.1,0.3,0.5],[0.2,0.4,0.6],[0.3,0.5,0.7]])
B1 = np.array([1,1,1])

print(W1.shape)
print(X.shape)
print(B1.shape)

A1 = np.dot(X,W1) + B1
Z1 = sigmoid(A1)
print(A1)
print(Z1)
```

```python
W2 = np.array([[0.2,0.4,0.6],[0.1,0.3,0.5],[0.4,0.6,0.8]])
B2 = np.array([1,1,1])

print(W2.shape)
print(B2.shape)

A2 = np.dot(Z1,W2) + B2
Z2 = sigmoid(A2)

print(A2)
print(Z2)
```

```python
W3 = np.array([[0.1,0.3],[-0.1,-0.5],[0.3,0.5]])
B3 = np.array([1,1])

print(W3.shape)
print(B3.shape)

A3 = np.dot(Z2,W3) + B3
Z3 = sigmoid(A3)

print(A3)
print(Z3)
```

```python
W4 = np.array([[0.1,0.2],[0.3,0.5]])
B4 = np.array([1,1])

print(W4.shape)
print(B4.shape)

A4 = np.dot(Z3,W4) + B4
Y = sigmoid(A4)

print(A4)
print(Y)
```

```python
def network():
    network = {}
    #첫번째 레이어
    network['W1'] = np.array([[0.1,0.3,0.5],[0.2,0.4,0.6],[0.3,0.5,0.7]])
    network['B1'] = np.array([1,1,1])
    
    #두번째 레이어
    network['W2'] = np.array([[0.2,0.4,0.6],[0.1,0.3,0.5],[0.4,0.6,0.8]])
    network['B2'] = np.array([1,1,1])
    
    #세번째 레이어
    network['W3'] = np.array([[0.1,0.3],[-0.1,-0.5],[0.3,0.5]])
    network['B3'] = np.array([1,1])
    
    #네번째 레이어
    network['W4'] = np.array([[0.1,0.2],[0.3,0.5]])
    network['B4'] = np.array([1,1])
    
    return network
```

```python
def forward(network,x):
    W1,W2,W3,W4 = network['W1'],network['W2'],network['W3'],network['W4']
    B1,B2,B3,B4 = network['B1'],network['B2'],network['B3'],network['B4']
    
    A1 = np.dot(x,W1) + B1
    Z1 = sigmoid(A1)  
    
    A2 = np.dot(Z1,W2) + B2
    Z2 = sigmoid(A2)
    
    A3 = np.dot(Z2,W3) + B3
    Z3 = sigmoid(A3)
        
    A4 = np.dot(Z3,W4) + B4
    Y = sigmoid(A4)
    
    return Y
```

## 신경망 추론 실행

```python
net = network()
x = np.array([0.3,1.3,-2.2])
y = forward(net,x)

print(y)
```
