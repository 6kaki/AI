## 1차원 배열

```python
import numpy as np
```

```python
# np.array 넘파이 어레이를 만들어주는 함수
x = np.array([1,2,3])
x.shape # 넘파이 어레이의 형태(원소 수)를 알려주는 함수
```

```python
x.dtype #넘파이 어레이가 어떤 데이터 타입으로 구성되어있는지 보영주는 함수
```

```python
type(x) # 파이썬 기본 함수로 numpy.ndarray임을 알 수 있다.
```

## 2차원 배열

```python
x = np.array([[1,2],[3,4]])
print(x.shape)
print(x.dtype)
print(x*10) #브로드 캐스트 덕분에 10이 2차원 배열로 형태가 바뀌어 각각 원소 곱을하여 결과가 나온다.
print(x+10) #브로드 캐스트 덕분에 10이 2차원 배열로 형태가 바뀌어 각각 원소 합을하여 결과가 나온다.
```

## 원소 접근

```python
x = np.array([[51,55],[14,19],[0,4]])
print(x[0])
print(x[0][1])
```

```python
for row in x:
    print(row)
```

```python
x = x.flatten() #1차원 배열로 변환
#[51,55,14,19,0,4]
x[np.array([0,2,4])]
```

```python
#특정 조건을 만족하는 원소만도 접근이 가능
x>15
```

```python
x[x>15]
```

## matplotlib.pyplot

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,6,0.1)
y = np.sin(x)

plt.plot(x,y)
plt.show()
```

```python
x = np.arange(0,6,0.1)
y1 = np.sin(x)
y2 = np.cos(x)
```

```python
plt.plot(x,y1,label='sin')
plt.plot(x,y2,linestyle='--',label='cos')
plt.title('Sin & Cos')
plt.legend()
plt.show()
```
